```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# PRIMER PUNTO

```python
#leemos la data original de movies
data = pd.read_csv("../first_projectDS/dataSet/movies.csv")
data
```

```python
#miramos los datos nulos de el dataset original
data.info()
```

```python
#sacamos el valor del 5% de las peliculas con mayor ingresos 
valor = data['gross'].quantile(0.95)
new_data = data[data['gross'] > valor]
new_data
```

```python
#creamos un nuevo dataframe con solo las columnas gross y budget
new_data = pd.DataFrame({'gross':new_data['gross'], 'budget': new_data['budget']})
new_data 
```

```python
#observamos cuantos datos NaN tiene budget
new_data['budget'].isnull().sum()
```

```python
#eliminamos las filas con datos NaN
new_data = new_data.dropna()
new_data.isnull().sum()
```

```python
#reordenamos los indidces
new_data = new_data.reset_index(drop=True)
```

```python
#sns.scatterplot(data=new_data,x='budget',y='gross')
sns.lmplot(x="budget", y="gross", data=new_data, line_kws={'color':'red'}, scatter_kws={'color':'blue'})
```

```python
new_data.describe()
```

# TERCER PUNTO

```python
#Cual es el escritor con mejor calificación
```

```python
conteo = data['writer'].value_counts()
conteo = conteo.head()
conteo_dataFrame = pd.DataFrame({'rating':conteo.index, 'quantity':conteo})
sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(x='rating',y='quantity', data=conteo_dataFrame)
plt.axhline(y=40)

```

```python
meanScore = []
for i in conteo.index:
  new = data[data['writer'] == i]
  mean= new['score'].mean()
  meanScore.append(mean)
meanScore
```

```python
series = pd.Series(meanScore)
series
```

```python
dataframe = pd.DataFrame(pd.Series(series),columns=['score'])
dataframe
```

```python
series2 = pd.Series(conteo.index)
series2
```

```python
dataframe2 = pd.DataFrame(pd.Series(series2),columns=['writer'])
dataframe2
```

```python
writer_dataframe = pd.DataFrame({'writer':dataframe2['writer'], 'score': dataframe['score']})
writer_dataframe 
```

```python
sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(x='writer',y='score', data=writer_dataframe)
plt.axhline(y=10)
```

# Quinto punto


```python
#Cual es el año de las peliculas con la mayor cantidad de votos promedio de los usuarios
```

```python
conteo = data['year'].value_counts()
conteo

conteo
```

```python
cantVotes = []
for i in conteo.index:
  yearData = data[data['year'] == i]
  acum= yearData['votes'].sum()
  cantVotes.append(acum)
cantVotes
```

```python
series = pd.Series(cantVotes)
series
```

```python
dataframe = pd.DataFrame(pd.Series(series),columns=['votes'])
dataframe
```

```python
series2 = pd.Series(conteo.index)
series2
```

```python
dataframe2 = pd.DataFrame(pd.Series(series2),columns=['year'])
dataframe2
```

```python
by_year = dataframe2.sort_values('year',ascending=True)
by_year
```

```python
yearVotes_dataframe = pd.DataFrame({'year':dataframe2['year'], 'votes': dataframe['votes']})
yearVotes_dataframe 
```

```python
sns.set(rc={'figure.figsize':(50,30)})
sns.barplot(x='year',y='votes', data = yearVotes_dataframe)
plt.axhline(y=10)
```
